```python
from bs4 import BeautifulSoup
from selenium import webdriver

options = webdriver.ChromeOptions()
options.add_experimental_option("detach", True)

nav = webdriver.Chrome(options=options)

url_base = "https://www.amazon.com.br/"

nav.get(url_base)

```

```python
from selenium.webdriver.common.keys import Keys

def procurar_produto():
    produto_nome = input('Qual produto deseja pesquisar? ')
    search = nav.find_element('xpath', '//*[@id="twotabsearchtextbox"]')
    search.clear()
    search.send_keys(produto_nome)
    search.send_keys(Keys.ENTER)
    
procurar_produto()
```

```python
site = BeautifulSoup(nav.page_source, 'html.parser')

produtos = site.findAll('div', attrs={"class": "a-section a-spacing-base"})

for produto in produtos:
    nome = produto.find('span', attrs={"class": "a-size-base-plus a-color-base a-text-normal"})

    preco = produto.find('span', attrs={"class": "a-offscreen"})

    parcela = produto.find('div', attrs={"class": "a-size-base a-color-secondary"})

    print(f'Título: {nome.text}')
    print(f'Preco: {preco.text}')
    if parcela:
        print(f'Parcela: {parcela.text}')
```
